```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
from numpy.linalg import norm
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib widget
```

```python
### read .dat and separete data 

# Search for ids that bind 
bind = []
with open ('dla.dat') as fid:
    for row in fid.readlines():
        if 'bind' in row:

            row = row.split()
            bind.append(int(row[1]))
        
print(bind)

# Separete move data of specific id
i = 1
f = str(bind[i])
x = []
y = []
z = []
with open ('dla.dat') as fid:
    for row in fid.readlines():
        if 'move' in row:
            #print(row)
            if f in row:
                row = row.split()
                x.append(int(row[2]))
                y.append(int(row[3]))
                z.append(int(row[4]))
```

```python
fig = plt.figure(figsize =(20,20))
ax = plt.axes(projection = '3d')
#ax.plot3D(x, y, z, 'blue')
xl = []
yl = []
zl = []
n = 10
for i in range(0,n):
    ax.scatter3D(x[i], y[i], z[i], color = "green")
    ax.text(x[i], y[i], z[i], ' %s, %s %s' % (str(x[i]), str(y[i]), str(z[i])), fontsize=12)

    xl.append(x[i])
    yl.append(y[i])
    zl.append(z[i])


ax.plot3D(xl, yl, zl, 'blue')
```
