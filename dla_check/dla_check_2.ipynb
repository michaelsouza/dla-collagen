```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
from numpy.linalg import norm
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib widget
```

```python
### read .dat and separete data 

# Search for ids that bind 
bind = []
with open ('dla.dat') as fid:
    for row in fid.readlines():
        if 'bind' in row:

            row = row.split()
            bind.append(int(row[1]))
        
#print(bind)

# Separete move data of specific id
i = 1
f = str(bind[i])
x = []
y = []
z = []
with open ('dla.dat') as fid:
    for row in fid.readlines():
        if 'move' in row:
            #print(row)
            if f in row:
                row = row.split()
                x.append(int(row[2]))
                y.append(int(row[3]))
                z.append(int(row[4]))
```
