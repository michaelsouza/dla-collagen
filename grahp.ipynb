```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,50,10)
y1 = x
y2 = x**2/10
y3 = 10*np.sin(x)
y4 = 50 - x
```

```python
plt.figure(figsize=(15,10), edgecolor='black', linewidth=3)


fontlabel = {'family': 'serif',
        'color':  'black',
        'weight': 'heavy',
        'style': 'italic',
        'size': 22,
        }

fontticks = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 20,}

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.xlabel('tempo(s)',fontdict=fontlabel)
plt.ylabel("Distancia(m)",fontdict=fontlabel)
plt.xticks(size = 19, family = 'serif', weight = 'normal')
plt.yticks(size = 19, family = 'serif', weight = 'normal')
#plt.yticks(fontdic = fontticks)

```
