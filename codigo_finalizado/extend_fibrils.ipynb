```python
import re
import os

def extract_ts_and_seed(filename):
    # Use express√µes regulares para extrair 'ts' e 'seed' do nome do arquivo
    match = re.match(r"dla_mode_s_ts_(\d+)_.*seed_(\d+)_.*\.dat", filename)
    if match:
        ts = match.group(1)
        seed = match.group(2)
        return ts, seed
    return None, None

def get_new_filename(filename):
    ts, seed = extract_ts_and_seed(filename)
    if ts and seed:
        return f"ts_{ts}_seed_{seed}.dat"
    return None
```
