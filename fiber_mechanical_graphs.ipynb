```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.font_manager as font_manager
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

fontlabel = {'family': 'serif',
        'color':  'black',
        'size': 24,
        }

fontlegend = font_manager.FontProperties(family='serif',
                                   style = 'italic', 
                                   size = 16)
```

```python
data1 = pd.read_csv('./porcet_parts_1.txt', delimiter='\t')
data2 = pd.read_csv('./porcet_parts_10.txt', delimiter='\t')
data3 = pd.read_csv('./porcet_parts_100.txt', delimiter='\t')
data4 = pd.read_csv('./porcet_parts_1000.txt', delimiter='\t')
data5 = pd.read_csv('./porcet_parts_10000.txt', delimiter='\t')

x1 = data1['0']
y1 = data1['100.0']
x2 = data2['0']
y2 = data2['100.0']
x3 = data3['0']
y3 = data3['100.0']
x4 = data4['0']
y4 = data4['100.0']
x5 = data5['0']
y5 = data5['100.0']
```

```python
x1 = [float(i) if(' ' not in i) else float(i[4:]) for i in x1.values]
```

```python
plt.plot(x1,y1)
plt.xlim(0,50)
plt.show()
```

```python
plt.figure(figsize=(12,12))

plt.plot(x1,y1, color = 'black', linewidth = 4,linestyle='dashed',markersize=12, marker = 's', label = r'$T_{s} = 1$')
plt.plot(x2,y2, color = 'firebrick', linewidth = 4,linestyle='dashed',markersize=12, marker = 'o', label = r'$T_{s} = 10$')
plt.plot(x3,y3, color = 'goldenrod', linewidth = 4,linestyle='dashed',markersize=12, marker = 'v', label = r'$T_{s} = 10^{2}$')
plt.plot(x4,y4, color = 'purple', linewidth = 4,linestyle='dashed',markersize=12, marker = 'D', label = r'$T_{s} = 10^{3}$')
plt.plot(x5,y5, color = 'darkgreen', linewidth = 4,linestyle='dashed',markersize=12, marker = '>', label = r'$T_{s} = 10^{4}$')
plt.legend( loc = 'best',prop=fontlegend)
plt.xlim(0,140)
plt.xlabel('Force applied to skeleton, F',fontdict=fontlabel)
plt.ylabel("% particles in skeleton",fontdict=fontlabel)
plt.xticks(size = 20, family = 'serif', weight = 'normal')
plt.yticks(size = 20, family = 'serif', weight = 'normal')
plt.tick_params(direction = 'in',bottom = True, top= True, left= True, right= True )

plt.show()
```
